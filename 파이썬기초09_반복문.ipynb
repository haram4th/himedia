# 반복문 for, while

반복문은 동일한 작업을 여러 번 반복해야 할 경우 사용하는 명령어<br>
반복문은 for와 while 두가지가 있는데
* for는 반복 횟수가 정해져 있는 경우
* while은 반복 횟수가 정해져 있지 않은 경우 사용

## 1. for : 반복할 횟수가 정해져 있는 경우

```
for 변수 in range(횟수):
    반복할 코드
```
```
for 변수 in 시퀀스 객체(list, tuple, 문자열):
    반복할 코드
```


* for문을 사용해서 1-100까지 숫자 출력

````python
for i in range(1, 101):
  print(i)
````

````python
for i in range(1, 101):
  print(i, end=" ")
````

````python
for i in range(1, 101):
  print(i, end=" ")
````

* for문을 사용해서 1-100 숫자 모두 더하기

````python
a = 10
````

````python
result = 0
for num in range(1, 101):
  result +=  num
print(result, end=" ")
````

* 입력한 횟수대로 반복하기

````python
count = int(input('반복할 횟수를 입력하세요: '))
# print(count)

for i in range(count):
  print(i+1, '번째 출력', 'Hello, Python')
````

## 시퀀스 객체로 반복하기(list, tuple, 문자열)
* 여러 개의 자료가 연속적으로 들어있는 list, tuple, 문자열 같은 시퀀스 객체를 for문에 사용하면 데이터를 한 개씩 꺼내오며 반복문이 실행된다.

리스트

````python
a = list(range(10, 101, 10))
a
````

````python
for i in a:
  print(i, end=" ")
````

````python
for i in "Hello, python":
  print(i)
````

역순으로 꺼내오기 reversed()

````python
for i in reversed('python loop'):
  print(i, end=" ")
````

````python
'python loop'[::-1]
````

````python
for i in 'python loop'[::-1]:
  print(i, end=" ")
````

````python
for i in a[::-1]:
  print(i, end=" ")
````

````python
for i in reversed(a):
  print(i, end=" ")
````

# 2. 반복할 횟수가 정해지지 않은 경우 while


```
변수 생성 및 초기화
while 조건식:
    반복할 코드
```



while 문을 사용해서 1-100 숫자 출력

````python
i = 1
while i <= 100:
  print(i, end=" ")
  i += 1
````

````python
i = 0
while i < 100:
  i += 1
  print(i, end=" ")
````

while 문을 사용해서 1-100까지 더하기

````python
i = 1
result = 0
while i <= 100:
  result += i
  i += 1
print(result, end=" ")
````

while 문을 이용해서 100-1까지 더하기

````python
i = 100
result = 0
while i >= 1:
  result += i
  print(result, end=" ")
  i -= 1
````

반복 횟수가 정해지지 않은 경우 while True 무한반복

````python
while True:
  print('while True로 무한 반복 출력')
````

True가 되는 경우: 1, 정수, 실수, 문자열, True, list, tutple, dictionary

````python
while [1,2,3]:
  print('while True로 무한 반복 출력')
````

````python
while {'key': 1}:
  print('while True로 무한 반복 출력')
````

False가 되는 경우: 0, False, None, 0.0, '', "", [], (), {}

````python
while []:
  print('while True로 무한 반복 출력')
````

break 문으로 반복문 탈출하기

````python
for i in range(1000000):
  print(i, end=" ")
  if i == 10:
    break
````

````python
i = 100
result = 0
while i >= 1:
  result += i
  print(result, end=" ")
  if i == 90:
    break
  i -= 1
````

````python
i = 1
while True:
  if i > 100:
    break
  print(i, end= " ")
  i += 1
````

continue로 코드 실행 건너뛰기

````python
for i in range(10):
  if i % 2 == 0:
    print(i, "짝수라서 숫자 인쇄 안하고 위로 올라감")
    print()
    continue
  else:
    print(i, "홀수라서 숫자 인쇄함")
  print("반복문 끝 print: ", i)
  print()
````

반복문에서 아무일도 하지 않고 건너뛰기: pass
* 우선 반복코드가 들어가야 하는 곳을 만들어 놓고 나중에 코딩하기 위해서 사용

````python
for i in range(10):
  pass
````

````python
# while은 pass가 통하지 않음
while True:
  pass
````

# 중첩 반복문
* 중첩 반복문은 for나 while문을 2번 이상 겹쳐서 사용하는 것
* 각 반복문의 블럭 위치를 조심
* 중첩 반복문의 실행 순서는 안쪽부터 바깥쪽으로 실행된다.

2중 중첩 반복문 예제
* 구구단 출력하기

````python
for i in range(1, 13):  # 2-9
  print("바깥쪽 for문 i: ", i,"시")
  for j in range(1, 60): # 1-9
    print("안쪽 for문 j:", j,"분")
    for k in range(1, 60): # 1-9
      print("안쪽 for문 k:", k, "초")
  print()
````

구구단

````python
for i in range(2, 10):
  for j in range(1, 10):
    print(f"{i} X {j} = ", i * j)
````

인덱스와 자료를 한번에 출력해주는 함수
enumerate(시퀀스객체(list, tuple, 문자열))

````python
a = range(1,11)
b = ["철수", "영희", '바둑이']
for index, value in enumerate(b):
  print(index, value)
````
