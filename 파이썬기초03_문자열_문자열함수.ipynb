* strip(): 문자열 양쪽 끝의 공백, 특수문자를 제거

```python
h12 = "     공백이 있는 문자열      "
h12
```

```python
h12.lstrip()
```

* rstrip(): 오른쪽 공백을 없애는 함수

```python
h12.rstrip()
```

* strip(): 양쪽 공백을 없애 줌

```python
h12.strip()
```

* .replace("찾을문자", "바꿀문자")

```python
h13
```

```python
h13.replace(".", "").replace(",", "").replace("!", "").replace("$", "").replace("%", "").strip()
```

```python
h13.replace("\n","").replace("\t","").replace(":", "").strip("., %!$")
```

```python
h13.translate(str.maketrans("","", "\n\t.:%$,!"))
```
