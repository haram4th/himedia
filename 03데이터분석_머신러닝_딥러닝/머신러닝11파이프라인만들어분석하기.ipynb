```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import koreanize_matplotlib
import seaborn as sns
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/haram4th/ADsP/main/wdbc.data', header=None)
data.head()
```

```python
data.info()
```

```python
data.loc[: , 1].value_counts()
```

```python
data[1] = data[1].apply(lambda x: 1 if x == 'M' else 0)
```

```python
data.head()
```

```python
X = data.drop([0,1], axis=1)
y = data[1]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=7)
```

파이프라인으로 정규화 => 차원축소PCA => 로지스틱 회귀 분석

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline
from sklearn.metrics import accuracy_score, classification_report
```

```python
pipe_lr = make_pipeline(StandardScaler(),
                        PCA(n_components=2),
                        LogisticRegression(random_state=7))
pipe_lr.fit(X_train, y_train)
pred = pipe_lr.predict(X_test)
print(accuracy_score(y_test, pred))
print(classification_report(y_test, pred))
                        
```
