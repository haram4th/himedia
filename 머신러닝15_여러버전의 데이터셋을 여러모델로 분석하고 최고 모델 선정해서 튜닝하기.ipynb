```python
X = data.drop(['G1','G2','G3'], axis=1)
y = data['G3']
```

```python
from sklearn.preprocessing import RobustScaler
rs = RobustScaler()
X_scaled = rs.fit_transform(X)
X_scaled = pd.DataFrame(X_scaled, columns=X.columns)
X_scaled.head(2)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=7)
```

```python
y
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
lr = LinearRegression()
lr.fit(X_train, y_train)
pred = lr.predict(X_test)
print(mean_squared_error(y_test, pred))
print(r2_score(y_test, pred))
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
rfr = RandomForestRegressor(n_estimators=5000, max_depth=5, n_jobs=-1)
rfr.fit(X_train, y_train)
pred = rfr.predict(X_test)
print(mean_squared_error(y_test, pred))
print(r2_score(y_test, pred))
```

```python
from imblearn.over_sampling import SVMSMOTE, ADASYN, SMOTE
svm = SVMSMOTE(random_state=7, n_jobs=-1)
ada = ADASYN(n_neighbors=2, random_state=7, n_jobs=-1)
smt = SMOTE(k_neighbors=2, random_state=7, n_jobs=-1)

# svm 
svm_X, svm_y = svm.fit_resample(X_train, y_train)

# ada
ada_X, ada_y = ada.fit_resample(X_train, y_train)

#smt
smt_X, smt_y = smt.fit_resample(X_train, y_train)
```

# SVM, RandomForest, XGBoost 

```python
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import RandomizedSearchCV
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score, classification_report
```

```python
# SMOTE  3  
dnames = ['SVM_SMOTE', 'ADA_SMOTE', 'SMOTE']
Xys = [[svm_X, X_test, svm_y, y_test], [ada_X, X_test, ada_y, y_test],
       [smt_X, X_test, smt_y, y_test]]
```

```python
#      
pipe_svm = Pipeline([('clf', SVC(random_state=7))])
pipe_rfc = Pipeline([('clf', RandomForestClassifier(random_state=7, n_jobs=-1))])
pipe_XGB = Pipeline([('clf', XGBClassifier(random_state=7, n_jobs=-1))])
pipe_MLP = Pipeline([('clf', MLPClassifier(random_state=7, max_iter=300))])
```

```python
randcv_params_svc = [dict(clf__kernel=['poly', 'rbf','sigmoid'])]
randcv_params_rfc = [dict(clf__max_depth=[3,5,7,10], clf__n_estimators=[100, 500, 1000],
                          clf__criterion=["gini", "entropy", "log_loss"])]
randcv_params_XGB = [dict(clf__max_depth=[3,5,7,10], clf__n_estimators=[100, 500, 1000],
                         clf__learning_rate=[0.01, 0.05, 0.1], 
                         clf__boosting_type=["dart", "gbdt"])]
randcv_params_MLP = [dict(clf__hidden_layer_sizes=[(100,), (150,),(200,)],
                         clf__activation=['identity', 'logistic', 'tanh', 'relu'],
                         clf__solver = ['lbfgs', 'sgd', 'adam'],
                         clf__learning_rate= ['constant', 'invscaling', 'adaptive'],
                         clf__learning_rate_init=[0.01, 0.05, 0.1])]
```

```python
clf_list = [pipe_svm, pipe_rfc, pipe_XGB, pipe_MLP]
param_list = [randcv_params_svc, randcv_params_rfc, randcv_params_XGB, randcv_params_MLP]
clf_name = {0 : "Support Vector Machine", 1 : "Random Forest", 2 : "XGBoost", 3 : "NeuralNet_MLP"}
```

```python
result_compare = []
for i, (dname, collist) in enumerate(zip(dnames, Xys)):
    for idx, (pipes, params) in enumerate(zip(clf_list, param_list)):
        rand_cv = RandomizedSearchCV(pipes, param_distributions=params, cv=3, n_jobs = -1, verbose=3)
        rand_cv.fit(collist[0], collist[2])
        pred = rand_cv.predict(collist[1])
        print(clf_name[idx], pred)
        print(clf_name[idx], sorted(rand_cv.cv_results_.keys()))
        print(clf_name[idx], ' : ', rand_cv.best_params_)
        print(clf_name[idx], f'  : {rand_cv.best_score_:.4f}')
        
        pipes.fit(collist[0], collist[2])
        pred = pipes.predict(collist[1])
        print(dname, clf_name[idx], accuracy_score(collist[3], pred), end="\n\n")
        result = {"Kind of data" : dname, "algorithm" : clf_name[idx],
              "params" : rand_cv.best_params_,
              "accuracy_score" : accuracy_score(collist[3], pred), # y_test
              "roc_auc_score" : roc_auc_score(collist[3], pred), # y_test
              "precision_score" : precision_score(collist[3], pred), # y_test
              "recall_score" : recall_score(collist[3], pred), # y_test
              "f1_score" : f1_score(collist[3], pred) # y_test
             }
        result_compare.append(result)
```

```python
result_df = pd.DataFrame()
for i in result_compare:
    temp = pd.DataFrame([i])
    result_df = pd.concat([result_df, temp])
result_df
```
