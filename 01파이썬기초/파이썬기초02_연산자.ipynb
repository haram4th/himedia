* 논리연산자: and, or, not

* and 비교하는 두 값 모두가 참이어야 참

```python
True and True
```

```python
a == b and b == a
```

```python
True and False
```

```python
a == b and a == c
```

```python
a > b and a == c
```

```python
False and False
```

```python
True or False
```

```python
False or True
```

```python
False or False
```

```python
a == b or a > b
```

```python
a < b or c < b
```
