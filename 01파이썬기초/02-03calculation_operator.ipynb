## 5) 식별연산자: is, is not  객체(object)가 같은지 비교

### A is B : A가 B와 같은지 비교 같으면 True, 다르면 False
### A is not B :  A와 B를 비교해서 다르면 True, 같으면 False
### ==, != 와 다른점 ==는 값 자체를 비교, is 와 is not은 객체(object)도 비교

```python
print(a)
print(b)
print(c)
```

```python
print(a == b)
print(a is b)
```

```python
print(a)
print(c)
print(a == c)
print(a is c)
print(type(a) is type(c))
```

### a(3)와 b(3.0)이 is로 비교 했을 때 다른 이유는 메모리 주소 값이 다르기 때문

```python
print(id(a))
print(id(b))
```

## 6) 논리연산자: and, or, not

### and는 두 값 모두가 참이어야 참

```python
True and True
```

```python
True and False
```

```python
False and True
```

```python
False and False
```

### or는 두 값 중 하나라도 참이면 참 두 값 모두가 거짓이어야 거짓

```python
True or True
```

```python
True or False
```

```python
False or True
```

```python
False or False
```

### not은 값을 반대로 뒤집음. True = False, False = True

```python
not True
```

```python
not False
```

### and, or, not 논리 연산자가 식 하나에 들어있으면 not, and, or 순으로 판단

```python
not True and False or not False
```

 위의 답이 True가 나오는 이유
 1. not True와 not False를 판단하여 False and False or True
 2. False and False를 판단하여 False
 3. False or True

```python
10 == 10 and 10 != 5    # True and True
```

```python
10 > 5 or 10 < 3        # True or False
```

```python
not 10 > 5              # not True
```

```python
not 1 is 1.0            # not False
```

## 7) 멤버연산자: in, not in

### in 객체 내에 해당 값이 있으면 True 없으면 False

```python
member = [1,2,3,4,5]
```

```python
1 in member
```

```python
7 in member
```

```python
mem = "in 객체 내에 해당 값이 있으면 True 없으면 False"
```

```python
'객체' in mem
```

```python
'파이썬' in mem
```

### not in 객체 내에 해당 값이 없으면 True 있으면 False

```python
7 not in member
```

```python
'파이썬' not in mem
```
