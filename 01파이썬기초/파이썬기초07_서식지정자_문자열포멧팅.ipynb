서식지정자나 format 메서드를 사용하는 이는 문자열로만 입력해야 하는 상황에서 같은 위치의 문자를 여러 번 바꿔야 할 때가 있기 때문이다.

# 1. 서식지정자 %d, %s, %f, %길이s

## 1) 문자를 입력할 때 %s

```python
print('이 수업의 이름은 %s 입니다.' % '머신러닝 빅데이터 분석')
```

```python
print('이 수업의 이름은 %s 입니다.' % '파이썬')
```

```python
subject = ['국어','영어','수학','과학','파이썬','장고']
```

```python
for sub_name in subject:
  print('이 수업의 이름은 %s 입니다.' % sub_name)
```

## 2) 정수를 입력할 때 %d

```python
print('올해는 %d년 입니다.' % 2023)
```

```python
year = [2021,2022,2023]
for y in year:
  print('올해는 %d년 입니다.' % y)
```

## 3) 실수를 입력할 때 %f

```python
print('사람의 정상 체온은 약 %f도 입니다.' % 36.5)
```

%f는 기본적으로 소숫점 6자리까지 표시하므로 f앞에 .자릿수를 지정하면 지정한 자릿수 만큼만 출력

```python
print('사람의 정상 체온은 약 %.2f도 입니다.' % 36.5)
```

```python
print('사람의 정상 체온은 약 %.1f도 입니다.' % 36.5)
```

## 4) 서식지정자로 문자열 정렬하기 %길이s

```python
print('%10s' % '오른쪽 정렬')
```

```python
print('%10d' % 19500)
```

숫자 오른쪽 정렬로 출력하기

```python
print('가격 %10d' % 1950000)
print('가격 %10d' % 71800)
```

```python
print('오늘의 삼성전자 주가는 %10d원 입니다.' % 71800)
```

## 여러 항목 동시에 넣기

```python
print('오늘은 %d년 %d월 %d일 입니다.' % (2021, 10, 6))
```

```python
company = ['삼성전자', 'LG전자','현대차','대한항공','하나투어']
stock = [71800,121000,195000, 31900, 86600]
```

```python
for com, sto in zip(company, stock):
  print('%4s의 주가는 %8d원 입니다.' % (com, sto))
```

# 2. format 메서드 사용하기
print('{인덱스}'.format(값))<br>
print('{인덱스0}{인덱스1}{인덱스2}'.format(값1, 값2, 값3))

```python
print('%s의 주가는 %d원 입니다.' % ('삼성전자', 71800))
print('{0}의 주가는 {1}원 입니다.'.format('삼성전자', 71800) )
print('{2}의 주가는 {1}원 입니다.'.format('삼성전자', 71800, '엘지전자') )
```

## format메서드로 같은 값을 여러 개 넣기

```python
print('{0}의 주가는 {1}원 입니다. \n{0}의 상한가는 {2},\n \
      {0}의 하한가는 {3}입니다.'.format('삼성전자', 71500, 50200, 93000))
```

## format 메서드에 변수로 값 넣기

```python
drama = '오징어 게임'
ranking = '1'
```

```python
print('넷플릭스 {}위 드라마는 {}입니다.'.format(ranking, drama))
```

```python
print(f'넷플릭스 {ranking}위 드라마는 {drama}입니다.')
```

```python
for com, sto in zip(company, stock):
  print(f'{com}의 주가는 {sto}원 입니다.')
```

## 5. format 메서트로 문자열 정렬하기

print('{인덱스:<길이}'.format(값))
print('{인덱스:방향 길이}'.format(값))

```python
# 오른쪽 방향으로 공백 포함 10칸
print('{0:<10}'.format('파이썬'))
```

```python
'{0:<10}'.format('파이썬')
```

```python
# 왼쪽 방향으로 공백 포함 10칸
print('{0:>10}'.format('파이썬'))
```

## 1) 숫자 갯수 통일하기
print('{인덱스:채울 숫자 길이 서식지정자}'.format(값), 단, 서식지정자에서 %는 빼고 적는다.

```python
print('{0:02d}'.format(2))
```

```python
# 실수로 채울 경우
print('{0:08.5f}'.format(2))
```

```python
month = range(1, 13)
day = range(1, 32)
```

```python
for mm in month:
  mon = '{0:02d}월'.format(mm)
  print(mon)
```
