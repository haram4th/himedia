## 3) 들여쓰기

### <li> 파이썬은 들여쓰기를 사용해서 코드를 구분해 작성한다.</li>
### <li> 들여쓰기를 잘 지키지 않으면 에러!!</li>

들여쓰기를 정상적으로 한 경우 (들여쓰기는 탭 / 공백4칸)

```python
a = 10
if a == 10:
    print('10입니다.')
```

들여쓰기를 지키지 않아서 오류가 나는 예 (IndentationError: expected an indented block)

```python
a = 10
if a == 10:
print('10입니다.')
```

오류는 나지 않지만 바르지 못한 예 (공백 5칸)

```python
a = 10
if a == 10:
     print('10입니다.')
```

오류는 나지 않지만 바르지 못한 예 (공백 이 너무 많음)

```python
a = 10
if a == 10:
                   print('10입니다.')
```

### <li> 들여쓰기 칸 수는 반드시 일치해야 함!!</li>

들여쓰기 칸 수가 일치하지 않아서 오류

```python
if a == 10:
    print('10')
  print('입니다.')
```

```python
if a == 10:
    print('10')
      print('입니다.')
```

```python
if a == 10:
    print('10')
    print('입니다.')
```

### 값을 여러 개 출력하기

print에는 변수나 값 여러 개를 ,(콤마)로 구분하여 넣을 수 있다.

print(값1, 값2, 값3)
print(변수1, 변수2, 변수3)

```python
a = 10 ; b= 20 ; c= 30
```

```python
print(a, b, c)
```

### sep로 값 사이에 문자 넣기

값 사이에 공백이 아닌 다른 문자를 넣고 싶을때 sep에 문자 또는 문자열을 지정 (sep는 구분자라는 뜻의 separator)

print(값1, 값2, sep='문자 또는 문자열')
print(변수1, 변수2, sep='문자 또는 문자열')

```python
print(1, 2, 3, sep=', ')    # sep에 콤마와 공백을 지정
```

```python
print(4, 5, 6, sep=',')    # sep에 콤마만 지정
```

```python
print('Hello', 'Python', sep='')    # sep에 공백을 지정
```

```python
print(1920, 1080, sep='x')    # sep에 x를 지정
```
