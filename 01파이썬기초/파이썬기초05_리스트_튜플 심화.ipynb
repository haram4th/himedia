리스트와 튜플, 문자열의 공통점은 연속된 값이 들어있다는 점.<br>
이런 연속적인 자료형을 시퀀스(sequence)자료형이라고 한다.

* 시퀀스 객체: list, tuple, string, range() 자료형

# 1. 특정 값이 시퀀스 객체 내에 있는지 확인하기
: in 시퀀스 객체(list, tuple, string, range())

```python
a = list(range(1,10))
a
```

```python
7 in a
```

```python
7 not in a
```

```python
s = "우리는 오늘 삼복 더위에서 파이썬을 공부하고 있습니다."
```

```python
"삼복" in s
```

# 2. 특정 값이 없는지 확인하기: 값 not in 시퀀스 객체

```python
"삼계탕" not in s
```
