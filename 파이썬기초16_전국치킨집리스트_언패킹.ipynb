```python
shop_info =[]
for i in data:
    for j in i['body']['items']:
        bizesId = j['bizesId']  # 가게번호
        bizesNm = j['bizesNm']  # 가게명
        ctprvnCd = j['ctprvnCd'] # 시도코드
        ctprvnNm = j['ctprvnNm'] # 시도명
        signguCd = j['signguCd'] # 시군구 코드
        signguNm = j['signguNm'] # 시 이름
        ldongCd = j['ldongCd']  # 법정동 코드
        ldongNm = j['ldongNm']  # 법정동 명
        lnoAdr = j['lnoAdr']   # 지번주소
        rdnmAdr = j['rdnmAdr']  # 도로명주소
        flrNo = j['flrNo']    # 층넘버
        hoNo = j['hoNo']     # 호실 번호
        lon = j['lon']      # 경도
        lat = j['lat']      # 위도
        shop_info.append([bizesId, bizesNm, ctprvnCd, ctprvnNm, signguCd, signguNm, ldongCd, ldongNm, lnoAdr, rdnmAdr, flrNo, hoNo, \
            lon, lat])
```

```python
len(shop_info)
```

CSV 파일로 저장

```python
colnames = ['bizesId', 'bizesNm', 'ctprvnCd', 'ctprvnNm', 'signguCd', 'signguNm', 'ldongCd', 'ldongNm', 'lnoAdr',\
             'rdnmAdr', 'flrNo', 'hoNo', 'lon', 'lat']
```

```python
colnames
```

```python
len(shop_info)
```

```python
shop_info.insert(0, colnames)
```

```python
import csv
```

```python
with open("전국치킨집리스트01.csv", 'w', newline="", encoding='cp949') as file:
    writer = csv.writer(file)
    writer.writerows(shop_info)
```
