# 2. 리스트 표현식(list comprehension) 사용하기

* list 안에 for 반복문과 if else 조건식을 한 번에 적는 방법
* 다소 복잡해 보일 수 있지만, 여러 줄의 코드를 1줄로 줄일 수 있고, for, if else 문을 쓸 때 보다 처리 속도가 빠르다.

```
[식 for 변수 in list]
```

````python
# 0부터 9까지 숫자를 생성하여 리스트 생성
L = []
for i in range(1, 10):
  L.append(i)
L
````

````python
참 if 조건식 else 거짓
````

````python
# list 표현식으로 1-9 리스트 만들기
a = [i for i in range(1, 11)]
a
````

리스트 표현식을 써야하는 이유

````python
%%timeit
a = []
for i in range(1, 100000000):
  a.append(i + 5)
print(a[0:10])
````

````python
%%timeit

b = [i + 5 for i in range(1,100000000)]
print(b[0:10])
````

````python
%%timeit
import numpy as np

i = np.arange(1, 100000000)
print(type(i))
c = i + 5
print(c[0:10])
````

# 파일 읽기, 파일쓰기
* open()
* with open()

## 1) open()
* 파일에 문자열을 쓰거나 읽을 때는 open 함수로 파일을 열어서 write메서드를 사용해 저장
```
변수명 = open("파일이름", 파일모드(w,r,bw, br)
변수명.write("저장할 내용")
변수명.close()
```
close로 꼭 파일을 닫아 주어야 함.

파일 만들고 내용 저장하기

````python
file = open('test.txt', 'w')
file.write("파일 만들기 연습 입니다.")
file.close()
````

만든 파일 읽어오기: 옵션만 w => r
```
변수1 = open("읽어올 파일명", 'r')
변수2 = 변수1.read()
변수1.close()
```

````python
file2 = open('test.txt', 'r')
my_file = file2.read()
file2.close()
print(my_file)
````

## with문으로 파일 자동으로 닫기

````python
with open('test_with.txt', 'w') as file3:
  file3.write("with open을 사용하면 파일을 자동으로 닫아주어 편리하다.")
````

````python
with open('test_with.txt', 'r') as file4:
  my_file2 = file4.read()
print(my_file2)
````
