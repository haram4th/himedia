```python
unique_book_list = book_list_data.drop_duplicates(['book_id'], keep='first')
```

* genres의 단어를 벡터화해서 코사인 유사도를 통해 비슷한 장르를 가진 행을 찾고
* 유사도가 높은 행을 기준으로 내림차순 정렬

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(ngram_range=(1,2))
tags_mat = cv.fit_transform(unique_book_list['tags'])
print(tags_mat)
```

```python
tags_mat.shape
```

코사인 유사도를 사용해 gener_mat의 유사도를 산출 후 genre_sim에 저장
* 비슷한 장르 영화를 찾기 위해 계산
* 코사인 유사도는 주로 문자, 문서의 유사도 계산에 사용됨

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
tags_sim = cosine_similarity(tags_mat, tags_mat)
print(tags_sim.shape)
print(tags_sim[:2])
```

```python
unique_book_list['tags'][:4]
```

```python
sorted_tags_sim = tags_sim.argsort()[:, ::-1]
```

```python
sorted_tags_sim 
```
