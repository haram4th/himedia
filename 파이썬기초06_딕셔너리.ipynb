# 5. dict()로 딕셔너리 만들기

## 1) dict(key1 = value1, key2 = value2) 형식으로 만들기
* 키 값에 '', ""를 쓰지 않음
* 정수, 실수 는 안되고 문자열 키만 가능


```python
d = dict(정수 = 100, 실수 = 13.5, 리스트 = [1,2,3,4], 튜플 = (1,2,), 딕셔너리 = {'딕트' : 400})
d
```

```python
type(d['정수'])
```

```python
d2 = dict(2 = 100, 3.14 = "원주율")
d2
```

## 2) dict([(key1, value1), (key2, value2)]) 형식으로 만들기
* 리스트 안에 (key, value)의 튜플을 나열해 만듬

```python
d2 = dict([('정수', 100), ('실수', 13.5), (100, 300)])
d2
```

## 3) dict(zip([key1, key2, key3], [value1, value2, value3])) 형식으로 만들기

* zip 함수를 사용해서 첫번째 리스트를 key로, 두 번째 리스트를 value로 값을 매칭해서 합침

```python
d3 = dict(zip(['문자열', '리스트', '튜플', '딕셔너리', '블린'], ['딕셔너리의 값', [1,2,3,4], (5,6,7,8), {'key' : 'value'}, True]))
d3
```

```python
d3['리스트'][2]
```

```python
d3['튜플'][3]
```

```python
d3['딕셔너리']['key']
```

# 딕셔너리 함수 사용하기

## 1) setdefault(key, value):
* 딕셔너리에 키와 값을 추가
* 단, 키만 지정하면 값은 None으로 자동 입력

```python
d3 = {'a': 10, 'b' : 20, 'c' : 30}
d3
```

```python
d3['d'] =
```

```python
d3.setdefault('d')
d3
```

```python
d3.setdefault('e', 50)
d3
```

```python
d3.setdefault(30, [1,2,3])
d3
```

* setdefault는 기존에 키가 있어서 키가 중복되는 경우 값이 덮어쓰기 되지 않는다.

```python
d3['a'] = 30
d3
```

```python
d3.setdefault('a', 10)
d3
```

## 2) update(key = value)
* 딕셔너리 안에 있는 키 값을 변경

```python
d3
```

```python
d3.update(a=10)
d3
```

```python
d3.update(f=60)
d3
```

* update 함수 사용시 key가 문자가 아니라 숫자형일 경우
* update({key:value})형태로 사용

```python
d3.update({70 : [5,6,7,8]})
```

```python
d3
```

* update(리스트), update(튜플)로 딕셔너리 key와 value 값 수정하기

update([[key1, value1],[key2, value2], [key3, value3]]<br>
update(((key1, value1),(key2, value2), (key3, value3))

```python
d3
```

```python
d3.update([['a', '십'], ['b', '이십'], ['c', '삼십']])
d3
```

```python
d3.update((('d', '사십'), ('e', '오십')))
d3
```

```python
d3.update(((30, 30), (70, (1,2,3))))
d3
```
