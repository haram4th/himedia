yyyymm 년월, sido 시도, resNm 관광지명, NatNum 총방문자수, ForNum 외국인 수

```python
tour_info = []
tour_info2 = []
for p_num in data2:
  year = p_num.get('yyyymm', 'unknown')
  sido = p_num.get('sido', 'unknown')
  p_name = p_num.get('resNm', 'unknown')
  net_num = p_num.get('NatNum', 0)
  for_num = p_num.get('ForNum', 0)
  tour_info.append([year, sido, p_name, net_num, for_num])
  tour_info2.append({'년월' : year, '시도' : sido, '관광지명' : p_name, '총방문자수' : net_num, '외국인수' : for_num})
```
